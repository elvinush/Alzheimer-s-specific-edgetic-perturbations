```python
#A function is created to give each edge a perturbation status based on its perturbation scores

def determine_perturbation_status(row):
    # Extract the relevant codes ignoring NaNs
    codes = row.iloc[1:-1].dropna().unique()

    # Determine the status based on the codes
    if set(codes) == {'11'}:
        return 'Non perturbed'
    elif set(codes) == {'01'}:
        return 'Strictly gained'
    elif set(codes) == {'10'}:
        return 'Strictly lost'
    elif set(codes) == {'01', '11'}:
        return 'Partly gained'
    elif set(codes) == {'10', '11'}:
        return 'Partly lost'
    elif set(codes) == {'10', '01'}:
        return 'Partly gained/lost'
    elif set(codes) == {'10', '11', '01'}:
        return 'Partly gained/lost'
    else:
        return np.nan  # In case no pattern is matched

# Apply the function to each row
merged['Perturbation Status'] = merged.apply(determine_perturbation_status, axis=1)

merged.head()
```
