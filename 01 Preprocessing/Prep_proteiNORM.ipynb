```python
#This script turns datasets into the necessary format to be read by ProteiNORM

import numpy as np
import pandas as pd
```

```python
#read and clean data
df = pd.read_csv(r"C:\Users\elvka\.Desktopreal\university\Thesis\Data\CSF\TMT\syn44132374\syn44132374_TMT.csv")

#df = df.drop(df.columns[0],axis=1)
df.head()
```

```python
df.rename({'protein IDs': 'id'}, axis=1, inplace=True)
df.head()
```

```python
names = list(df)
names.remove(names[0])
print(names)
```

```python
names2 = []
for i in names:
    x = i.replace('TMT1_b', '')
    x = x.replace('.', ' ')
    x = 'Reporter intensity corrected ' + x
    names2.append(x)



print(names2)
```

```python
renamed = dict(zip(names, names2))

print(renamed)
```

```python
df.rename(renamed, axis=1, inplace=True)
df.head()
```

```python
df2 = df.iloc[:len(df.columns),:]
df2.shape
```

```python
#This is the final format that datasets need to be evaluated by ProteiNORM

df2.head()
```

```python
df2.to_csv(r"C:\Users\elvka\.Desktopreal\university\Thesis\Data\CSF\TMT\syn44132374\syn44132374_TMT_prep_rect.csv", index=False, sep='\t')
```
