```python
#This code averages out the abundance level for both the diseased and the healthy group

import numpy as np
import pandas as pd
```

```python
#Read the imputed, diseased dataset

df_ad = pd.read_csv(r"C:\Users\elvka\.Desktopreal\university\Thesis\Data\Brain\TMT\Rosmap_r1\Rosmap_r1_imp_ad.csv")

df_ad.head()
```

```python
#Read the imputed, healthy dataset

df_ctl = pd.read_csv(r"C:\Users\elvka\.Desktopreal\university\Thesis\Data\Brain\TMT\Rosmap_r1\Rosmap_r1_imp_ctl.csv")

df_ctl.head()
```

```python
#Average the abundance data of each protein across all diseased samples

df_ad_means = df_ad.loc[:, df_ad.columns != 'protein IDs'].mean()


df_ad_means = pd.DataFrame(df_ad_means, columns = ['Abundance_AD'])
df_ad_means = df_ad_means.reset_index().rename(columns={'index': 'Official Symbol Interactor'})

df_ad_means.head()
```

```python
#Explode the dataset, so each protein name has its own designated row. Otherwise there would be rows that have multiple synonym protein names assigned to them.

df_ad_means['Official Symbol Interactor'] = df_ad_means['Official Symbol Interactor'].str.split('|')

df_ad_means = df_ad_means.explode('Official Symbol Interactor').reset_index(drop=True)

df_ad_means.head()
```

```python
print(len(df_ad_means))
```

```python
#Average the abundance data of each protein across all healthy samples

df_ctl_means = df_ctl.loc[:, df_ctl.columns != 'protein IDs'].mean(axis=0)

df_ctl_means = pd.DataFrame(df_ctl_means, columns = ['Abundance_CTL'])
df_ctl_means = df_ctl_means.reset_index().rename(columns={'index': 'Official Symbol Interactor'})

df_ctl_means.head()
```

```python
#Explode the dataset, so each protein name has its own designated row. Otherwise there would be rows that have multiple synonym protein names assigned to them.

df_ctl_means['Official Symbol Interactor'] = df_ctl_means['Official Symbol Interactor'].str.split('|')

df_ctl_means = df_ctl_means.explode('Official Symbol Interactor').reset_index(drop=True)

df_ctl_means.head()
```

```python
print(len(df_ctl_means))
```
